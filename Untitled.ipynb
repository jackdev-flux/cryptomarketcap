```python
# Fetch JSON data via Nomics API
import urllib.request

url = "https://api.nomics.com/v1/market-cap/history?key=94cd28819a4946946ae3118c7df9ea20613f191a&start=2013-04-14T00%3A00%3A00Z&end=2021-09-14T00%3A00%3A00Z"
print(urllib.request.urlopen(url).read())
```

```python
# Fetch again, and convert to DataFrame for querying and transformation
import pandas as pd
import urllib.request

url = "https://api.nomics.com/v1/market-cap/history?key=94cd28819a4946946ae3118c7df9ea20613f191a&start=2013-04-14T00%3A00%3A00Z&end=2021-09-14T00%3A00%3A00Z"
jsonData = urllib.request.urlopen(url).read()
df = pd.read_json(jsonData)
print(df)
```

```python
# Write the output to a csv for excel analysis
df.to_csv('output.csv')
```

```python
# Visualize in a line chart using MatPlotLib
import matplotlib.pyplot as plt 

plt.rcParams["figure.figsize"] = (15, 6)

df.plot.line(x='timestamp',y='market_cap',color='green')
plt.show()
```
